```
import pandas as pd
df = pd.read_csv('Tweets.csv')
```

```
new = df.drop(columns=['airline_sentiment_gold', 'retweet_count', 'airline_sentiment', 'tweet_coord','negativereason_gold'])

new
```

```
from textblob import TextBlob
from wordcloud import WordCloud
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
```

```
new.head()
```

```
new.describe()
```

```
def cleanTxt(text):
 text = re.sub('@[A-Za-z0–9]+', '', text) 
 text = re.sub('#', '', text) 
 text = re.sub('RT[\s]+', '', text) 
 text = re.sub('https?:\/\/\S+', '', text) 
 
 return text



new['text'] = new['text'].apply(cleanTxt)
new
```

```
def getSubjectivity(text):
   return TextBlob(text).sentiment.subjectivity


def getPolarity(text):
   return  TextBlob(text).sentiment.polarity



new['Subjectivity'] = new['text'].apply(getSubjectivity)
new['Polarity'] = new['text'].apply(getPolarity)
new
```

```
def getAnalysis(score):
  if score < 0:
    return 'Negative'
  elif score == 0:
    return 'Neutral'
  else:
    return 'Positive'
    
new['Analysis'] = new['Polarity'].apply(getAnalysis)

new
```
